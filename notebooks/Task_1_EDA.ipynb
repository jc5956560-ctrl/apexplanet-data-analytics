```python
import os
import pandas as pd
path1 = os.path.expanduser("~/Downloads/NGO Data Analytics - Task 1.xlsx")
path2 = os.path.expanduser("~/Onedrive/Downloads/NGO Data Analytics -Task 1.xlsx")
file_path = path1 if os.path.exists(path1) else path2
df = pd.read_excel(file_path)
df.head()
```

```python
print("Dataset Shape:", df.shape)
print("\nMissing Vlues per Column:")
(df.isnull().sum())
```

```python
print(df.columns.tolist())
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, y='Area of work', order=df['Area of work'].value_counts().index, palette='viridis')
plt.title('Distribution of NGOs by Area of Work')
plt.xlabel('Count of NGOs')
plt.ylabel('Area of Work')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 6))
df['Headquaters'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('NGOs by Headquarters Location')
plt.ylabel('')
plt.show()
```
